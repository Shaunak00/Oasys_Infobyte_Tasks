```python
import pandas as pd

# Load the data
file_path1 = 'Unemployment in India.csv'
file_path2 = 'Unemployment_Rate_upto_11_2020.csv'

unemployment_india = pd.read_csv(file_path1)
unemployment_rate = pd.read_csv(file_path2)

# Display the first few rows of each dataset
unemployment_india.head(), unemployment_rate.head()
```

```python
# Check for missing values and data types
unemployment_india.info()
unemployment_rate.info()

# Check for missing values in each dataset
unemployment_india.isnull().sum(), unemployment_rate.isnull().sum()
```

```python
# Drop rows with missing values in the first dataset
unemployment_india_clean = unemployment_india.dropna()

# Convert the 'Date' columns to datetime format
unemployment_india_clean['Date'] = pd.to_datetime(unemployment_india_clean[' Date'])
unemployment_rate['Date'] = pd.to_datetime(unemployment_rate[' Date'])

# Drop original date columns with extra spaces
unemployment_india_clean.drop(columns=[' Date'], inplace=True)
unemployment_rate.drop(columns=[' Date'], inplace=True)

# Display the cleaned datasets
unemployment_india_clean.head(), unemployment_rate.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting the unemployment rate over time for the first dataset
plt.figure(figsize=(14, 7))
sns.lineplot(data=unemployment_india_clean, x='Date', y=' Estimated Unemployment Rate (%)', hue='Region')
plt.title('Unemployment Rate Over Time (Dataset 1)')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.legend(loc='upper right')
plt.show()

# Plotting the unemployment rate over time for the second dataset
plt.figure(figsize=(14, 7))
sns.lineplot(data=unemployment_rate, x='Date', y=' Estimated Unemployment Rate (%)', hue='Region')
plt.title('Unemployment Rate Over Time (Dataset 2)')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.legend(loc='upper right')
plt.show()
```

```python
# Summary statistics for the first dataset
print("Summary Statistics for Dataset 1:")
print(unemployment_india_clean.describe())

# Summary statistics for the second dataset
print("Summary Statistics for Dataset 2:")
print(unemployment_rate.describe())
```

```python
# Correlation matrix for the first dataset
print("Correlation Matrix for Dataset 1:")
print(unemployment_india_clean.corr())

# Correlation matrix for the second dataset
print("Correlation Matrix for Dataset 2:")
print(unemployment_rate.corr())
```

```python
# Average unemployment rate by region for the first dataset
print("Average Unemployment Rate by Region (Dataset 1):")
print(unemployment_india_clean.groupby('Region')[' Estimated Unemployment Rate (%)'].mean())

# Average unemployment rate by region for the second dataset
print("Average Unemployment Rate by Region (Dataset 2):")
print(unemployment_rate.groupby('Region')[' Estimated Unemployment Rate (%)'].mean())
```

```python
# Moving average for the first dataset
unemployment_india_clean['Moving_Avg'] = unemployment_india_clean[' Estimated Unemployment Rate (%)'].rolling(window=3).mean()

plt.figure(figsize=(14, 7))
sns.lineplot(data=unemployment_india_clean, x='Date', y='Moving_Avg', hue='Region')
plt.title('3-Month Moving Average of Unemployment Rate Over Time (Dataset 1)')
plt.xlabel('Date')
plt.ylabel('3-Month Moving Average Unemployment Rate (%)')
plt.legend(loc='upper right')
plt.show()

# Moving average for the second dataset
unemployment_rate['Moving_Avg'] = unemployment_rate[' Estimated Unemployment Rate (%)'].rolling(window=3).mean()

plt.figure(figsize=(14, 7))
sns.lineplot(data=unemployment_rate, x='Date', y='Moving_Avg', hue='Region')
plt.title('3-Month Moving Average of Unemployment Rate Over Time (Dataset 2)')
plt.xlabel('Date')
plt.ylabel('3-Month Moving Average Unemployment Rate (%)')
plt.legend(loc='upper right')
plt.show()
```
